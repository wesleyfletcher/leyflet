```python
# import pandas as pd
from threading import Thread
import os

from scrapers import *
```

```python
os.mkdir('TEMP_TABLES')
```

```python
funcs = [appoll, kenpom, net, torvik]
```

```python
threads = [Thread(target=func) for func in funcs]
```

```python
for t in threads:
    t.start()

for t in threads:
    t.join()
```

```python
data_frames = [pd.read_csv(f'TEMP_TABLES/{file}', index_col='Team').rename(columns={'Rank' : file[:-4]}) for file in os.listdir('TEMP_TABLES')]
for df in data_frames:
    df.index = df.index.map(get_name)
    df[df.columns[0]] = df[df.columns[0]].astype(str)

table = data_frames[0]
for i in range(1, len(data_frames)):
    table = pd.merge(table, data_frames[i], how='outer', left_index=True, right_index=True)

table.reset_index().to_csv('composite.csv', index=False)
```
